# Import required libraries

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
```

# Load dateset

```python
iris = load_iris() # Load features X and labels Y
X = iris.data      # Feature variables
Y = iris.target    # Target variables (species of iris)
```

# Split dataset into training and testing sets

```python
# 70% for training and 30% for testing
# random state ensures reproducibility
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

# Create and train model

```python
clf = DecisionTreeClassifier()
clf.fit(X_train, Y_train)
```

# Make prediction and evaluate model

```python
Y_pred = clf.predict(X_test)
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

# Visualize the decision tree

```python
plt.figure(figsize=(12, 8))
plot_tree(
    clf,
    filled = True, # Color of nodes by class
    feature_names = iris.feature_names,
    class_names = iris.target_names
)

plt.title("Decision Tree Classifier for Iris Dataset")
plt.show()
```
